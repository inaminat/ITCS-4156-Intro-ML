### 1) Load the Diabetes dataset into a dataframe (10 Marks)

```python

# Your input goes here
from sklearn.datasets import load_diabetes

diabetes_data=load_diabetes()
print(diabetes_data.DESCR)
```

```python
# The target column of this dataset is not named unlike MEDV in Boston housing dataset. 
# Therefore, you can rename it as Output to use it in the dataset 
import pandas as pd


# Your input goes here
df=pd.DataFrame(data=diabetes_data.data, columns=diabetes_data.feature_names)
df["Output"]=diabetes_data.target
df
```

### 2) Create your own Training and Testing data and print them out (10 Marks)

```python

# Your input goes here
from sklearn.model_selection import train_test_split

X = df.drop(['Output'], axis=1)
Y = df['Output']

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2, random_state=1)
```

```python
# Print x_test
x_test
```

### 3) Fit a model using a single feature. Use 'BMI' as the selected feature (10 Marks)

```python


# Your input goes here
from sklearn import linear_model   # Defines regression model 
model = linear_model.LinearRegression()   # Build a training model
model.fit(x_train['bmi'].values.reshape(-1,1), y_train)
```

### 4) Predict the data, print it and compare it with the original (10 Marks)

```python


# Apply trained model to make prediction on test set 

y_pred = model.predict(x_test['bmi'].values.reshape(-1,1))     #Apply trained model to make prediction on test set 
```

```python
#I am going to use numpy array instead DataFrame, because I got the error, that says that boundaries were wrong
#DataFrame series has to be inverted to numpy. So, I decided don't do cast to DataFarme and convert y_pred itself

# Print the Predicted data
y_pred = y_pred.to_numpy()
y_pred
```

```python


# Compare y_test and y_pred

y_test,y_pred
```

### 5) Plot the data (10 Marks)

```python

# Your input goes here
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(8,5))



# label axes 
plt.xlabel(" BMI ") 
plt.ylabel(" Output ")

# Scatter plot actual test set values in Black

plt.scatter(x_test["bmi"], y_test, color ='black')

# line plot for predictions 
plt.plot(x_test["bmi"],  y_pred, color = "blue", linewidth = 3)

# plotting the graph
```

### 6) Print the Model Performance (10 Marks)

```python


# Your input goes here
from sklearn.metrics import mean_squared_error, r2_score

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(y_test, y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(y_test, y_pred))
```
